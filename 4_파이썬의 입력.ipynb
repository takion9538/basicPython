# **1. input 함수**

* 사용자에게 데이터를 입력받을 수 있는 함수
* 데이터를 문자형으로 입력받음

```python
name = input()
print(name)
```

```python
print('이름을 입력하세요 : ', end='')
name = input()
print('%s 님 안녕하세요!' % name)
```

* 인수 : 입력받은 변수 

```python
# split() : 인수로 전달된 문자를 기준으로 데이터를 나눠서 전달
print('이름과 나이를 입력하세요 : ', end='')
name, age, height = input().split(',')
print('이름 : {0}, 나이 : {1}, 키 : {2}'.format(name, age, height))
```

```python
# split() : 인수로 전달된 문자를 기준으로 데이터를 나눠서 전달
print('이름과 나이를 입력하세요 : ', end='')
name, age, height = input().split('※')
print('이름 : {0}, 나이 : {1}, 키 : {2}'.format(name, age, height))
```

```python
kor, eng, math = input('시험 점수를 입력하세요: ').split(', ')
print(kor + eng + math)
print(type(kor), type(eng), type(math))

# int() : 문자열을 정수 타입으로 변환
# float() : 문자열을 실수 타입으로 변환
# str() : (모든)자료형을 문자 타입으로 변환
print(int(kor) + int(eng) + int(math))

# print(int(kor + eng + math)) # -> 908070 의 정수형으로 변환됨
```

# **2. map 함수**

* 미리 타입을 지정하여 입력받는 함수

```python
kor, eng, math = map(int, input('시험점수를 입력하세요 : ').split(', '))
print(kor + eng + math)
```
