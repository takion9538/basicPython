# **1. 상속**

* 기존의 정의해둔 클래스의 기능을 그대로 물려받음
* 기존 클래스에 기능 일부를 추가하거나, 변경하여 새로운 클래스를 정의함
* 기존의 안정적인 코드를 재사용할 수 있게 됨
* 상속을 받고자 하는 대상인 기존 클래스를 부모, parent, super, base 클래스라고 부름
* 상속을 받은 새로운 클래스는 자식, child, sub 클래스 라고 부름

```python
class Animal :
  def __init__(self, name, age) : 
    self.name = name
    self.age = age

  def eat(self, food) :
    print('{} 는 {} 를 먹습니다.'.format(self.name, food))

  def sleep(self, hour) :
    print('{} 는 {} 시간 동안 잡니다.'.format(self.name, hour))
```

```python
animal = Animal('동물', 6)
animal.eat('먹이')
animal.sleep(5)
```

```python
class Dog(Animal) : # Animal 클래스를 상속받음
  pass
```

```python
Rucy = Dog('Rucy', 6)
Rucy.eat('사료')
Rucy.sleep(7)
```

# **2. 메소드 오버라이딩**

* 부모 클래스의 method 를 재정의(override)
* 자식 클래스의 인스턴스로 호출시 , 재정의된 메소드가 호출됨

```python
class Animal :
  def __init__(self, name, age) : 
    self.name = name
    self.age = age

  def eat(self, food) :
    print('{} 는 {} 를 먹습니다.'.format(self.name, food))

  def sleep(self, hour) :
    print('{} 는 {} 시간 동안 잡니다.'.format(self.name, hour))
```

```python
class Dog(Animal) :
  def eat(self, food) :
    print('{} 는 {} 를 맛있게 먹습니다.'.format(self.name, food)) # override

  def super_eat(self, food) :
    super().eat(food)
```

```python
animal = Animal('동물', 6)
animal.eat('먹이')
animal.sleep(5)
```

```python
Rucy = Dog('Rucy', 6)
Rucy.eat('사료')
Rucy.sleep(7)
Rucy.super_eat('밥')
```
