# **1. 객체지향 프로그래밍**

* 문제를 여러 개의 객체 단위로 나누어 작업하는 방식
* 클래스를 이용하여 연관있는 처리부분과 데이터 부분을 하나로 묶어 객체를 생성해 사용함

### **1-1. 클래스와 객체**

* 객체는 클래스로 생성되어 구체화된 인스턴스  
  (강아지, 자동차, 성적 등 머릿속으로 그려지는데 크게 두 가지로 표현할 수 있다 (1.명사(변수):(색:노랑, 나이:1살) 2.동사(함수):(뛴다, 짖는다 등)) 이와 같이 여러가지 특징이 같은 집합체를 클래스 라고 한다(자동차의 설계도 = 클래스, 아직 만들어지지 않은 틀을 정의해 놓은 것) 클래스를 토대로 제작하여 사용할 수 있는 형태를 객체라고 한다
  (파일의 형태로 존재하면 클래스, 메모리의 형태로 존재하면 객체))
* 실제로 클래스가 인스턴스화 되어 메모리에 상주하고 있는 상태를 객체라고 부른다
* 건축 설계도를 클래스로 비유하면, 실제로 지어진 집을 객체로 비유할 수 있다
* **파이썬의 모든 변수는 객체이다**

### **1-2. 클래스 만들기**

```
class 클래스명 :
  def __init__() :
    객체가 메모리에 올라갈 때 가장 먼저 실행될 문장
    
  def 메소드명 :
    메소드가 호출되면 실행될 문장
```

````python
# 기능이 없는 클래스 만들기

class Dog : # 클래스 선언
  pass # 내용이 없는 블록을 실행 (내용이 없으면 에러가 발생)
````

````python
# 클래스를 통해 객체를 생성하기

Rucy = Dog() # Dog 의 형태를 갖는 객체 Rucy 가 생성
PPomi = Dog()

print(type(Rucy)) # <class '__main__.Dog'> : Dog 의 설계도를 갖는 객체 Rucy 가 생성된 것을 확인
print(Rucy, PPomi) # 메모리의 주소가 출력됨(두 객체의 주소가 서로 다름 : 도그 설계도로 생성된 두 객체가 서로 다른 곳에 저장되어 있음을 확인)
````

````python
li1 = list([1, 2])
li2 = list([1, 2])
print(type(li1), type(li2)) # List 의 형태로 만들어진 객체
````

### **1-3. 생성자**

* 클래스 인스턴스가 생성될 때 호출됨(객체를 생성되고 나서 가장 먼저 실행되는 문장)
* __ init __(self) : 의 형태를 가짐
* self 매개변수는 항상 첫번째 오며, 자기 자신을 가리킴(자기 객체를 가리킴)
* 이름이 꼭 self 일 필요는 없지만, 관례적으로 self 를 사용
* 생성자에서는 해당 클래스가 다루는 데이터를 정의

````python
class Dog :
  def __init__(self) : # 이니셜라이즈, 초기화한 후 다음을 실행
    print(self, '생성됨')
    self.name = '무명' # self(자신) 에 대해 설명하는 기본값
    self.age = 0 
````

````python
Rucy = Dog() # 메모리에 객체가 생성되면서, 생성자(이니셜라이즈)가 실행됨
print(Rucy.name, Rucy.age) # Rucy 객체의 name 컴포넌트를 출력함

PPomi = Dog()
print(PPomi.name, PPomi.age)
````

````python
Rucy.name = 'Rucy'
Rucy.age = 12

print('객체 Rucy의 이름은 {} 이고, 나이는 {} 살 입니다.'.format(Rucy.name, Rucy.age))
print('객체 PPomi의 이름은 {} 이고, 나이는 {} 살 입니다.'.format(PPomi.name, PPomi.age))
````

````python
class Dog :
  def __init__(self, name, age, family='포메라니안') : # default 매개변수 = 포메
    self.name = name # 위의 self.name(field) = '무명' 에서 '무명' 부분에 init 의 name 이 저장됨
    self.age = age
    self.family = family
````

````python
Rucy = Dog('Rucy', 12, '포메라니안') # 객체를 생성할 때 name, age 를 정의하지 않으면 생성되지 않음
print(Rucy.name, Rucy.age, Rucy.family)

PPomi = Dog('PPomi', '6', '닥스훈트')
print(PPomi.name, PPomi.age, PPomi.family)
````

### **1-4. 메소드 정의하기**

* 멤버함수라고도 하며, 해당 클래스의 객체에서만 호출가능한 함수를 의미함
* 메소드는 객체에서만 호출되며, 해당 객체의 속성에 대한 연산을 행할 때 쓰임
* 객체.메소드명() 형태로 호출할 수 있다

````python
class Counter :
  def __init__(self) :
    self.num = 0  
  def increment(self) : 
    self.num += 1
  def current_value(self) : 
    return '대기인원 : ' + str(self.num)
  def reset(self) : 
    self.num = 0
````

````python
SHbank = Counter()
print(SHbank.num)

SHbank.increment()
print(SHbank.num)

SHbank.current_value()
````

````python
HanaBank = Counter()
print(HanaBank.num)

HanaBank.increment()
HanaBank.increment()
HanaBank.increment()
HanaBank.increment()
print(HanaBank.num)

print(HanaBank.current_value())

HanaBank.reset()
print(HanaBank.num)
````

### **1-5. 메소드 타입**

* instance method : 객체 형태로 호출되기 때문에 해당 메소드를 호출한 객체에만 영향을 미침
* class method : class 로 호출됨(객체를 생성하지 않아도 사용 가능)
> 함수 선언 위에 @staticmethod 로 표기하여 생성함

````python
class Math :
  def add(self, x, y) :
    return x + y
  def multiply(self, x, y) :
    return x * y
````

````python
math = Math()
result1 = math.add(10, 5) # 인스턴트 메소드(객체를 생성해야만 사용할 수 있음)
result2 = math.multiply(10, 5)
print(result1, result2)
````

````python
class Math :
  @staticmethod # 객체를 생성하지 않아도 사용할 수 있도록 생성함
# def add(self, x, y) : # self 를 사용하면 안된다 -> self 는 자신의 객체를 가리키기 때문
  def add(x, y) :
    return x + y
  @staticmethod # 객체를 생성하지 않아도 메모리에 저장됨, 따라서 바로 사용할 수 있음
  def multiply(x, y) :
    return x * y
````

````python
result1 = Math.add(10, 5)
print(result1, Math.multiply(10, 5))
````
