# **1. 오류처리**

* 오류가 발생되었을 때 별도의 처리를 하거나 무시하고 프로그램을 실행하는 것
* 오류가 발생될 것으로 예상되는 문장을 try 블록에 코딩하고, try 블록 안에 코딩된 내용을 실행하여 오류가 발생할 경우 더이상 try 블록의 내용을 실행하지 않고, 해당 오류의 except 블록에 코딩된 내용을 실행
```
try :
  오류가 발생될 것으로 예상되는 문장
  ...
except [오류이름:]
  오류가 발생되면 실행할 문장
  ...
except [오류이름2:]
  오류가 발생되면 실행할 문장
  ...
except [오류이름3:]
  오류가 발생되면 실행할 문장
  ...
finally :
  오류 발생 여부와 관계없이 무조건 실행해야할 문장(있다면 작성)
  ...
```

* [] 안의 오류 이름은 생략할 수 있음
* finally 는 선택사항

````python
try :
  print(10 / 2)

except :
  print('오류가 발생했습니다')

print('실행 완료')
````

````python
print(10 / 0)
print('오류가 발생했습니다')
````

````python
try :
  print(5 / 2)
  # ZeroDivisionError
  print(10 / 0) # 오류가 발생되면 오류가 발생된 문장부터 나머지 문장을 실행하지 않고 except로 이동
  print(10 / 5) # 오류가 발생할 경우 이 문장은 실행되지 않음

except ZeroDivisionError :
  print('오류가 발생했습니다')

finally :
  print('0 으로는 나눌 수 없습니다')

print('실행 완료')
````

````python
data = [10, 20, 30]

try :
  print(data[2])
  # IndexError
  print(data[3]) # 실행하지 않고 except로 이동

except :
  print('오류가 발생했습니다')
print('실행 완료')
````
