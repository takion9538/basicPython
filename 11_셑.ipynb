# **1. Set**

* 수학의 집합과 비슷한 형태로 순서가 없기 때문에 어떤 값이 먼저 나올지 알 수 없고, 중복 데이터를 허용하지 않음 (= 인덱스가 없음)
* 딕셔너리처럼 중괄호를 사용해서 선언하지만, key 는 존재하지 않고 value 만 존재하는 형태

### **1-1. Set 만들기**

```python
s1 = set()
print(s1)

s2 = set([1, 3, 5, 7])
print(s2)

s3 = {1, 3, 5, 7}
print(s3)

s4 = {1, 1, 3, 5, 5, 3, 5, 7, 9}
print(s4) # 중복된 값은 1개만 적용됨
```

```python
print(2 in s4) # s4 에서 2 를 찾는 명령
print(5 in s4)
print(3 not in s4) # 논리연산자의 앞에
```

### **1-2. Set 함수**

```python
# add() : set 에 단일 데이터를 추가
s1 = set([100, 200])
s1.add(150)
print(s1) # 순서가 없는 데이터 형식
s1.add(20)
print(s1)

# update() : set 에 여러 데이터를 한번에 추가
s1.update([10, 20, 30, 40]) # 파이썬에서 데이터 묶음을 받아들이는 방식 []
print(s1)

# remove() : set 의 데이터를 제거, 제거할 데이터가 없으면 에러가 발생
s1.remove(40)
print(s1)
# s1.remove(40) # KeyError : 40 (데이터가 없음)

# discard() : set 의 데이터를 제거, 제거할 데이터가 없어도 에러가 발생하지 않음
s1.discard(10)
print(s1)
s1.discard(10) # 에러가 발생하지 않음
print(s1)
# 굳이 remove 와 discard 가 같이 존재하는 이유? : discard 는 데이터가 사라지기만 하면 상관 없을 때 사용, remove 는 없는 메시지를 지웠을 때 발생하는 에러등을 다른 프로그램 또는 로직으로 전송해 실행하는 예외처리를 할 때 사용됨 

# copy() : set 을 복사
s10 = s1.copy()
print(s1)
print(s10)
s10 = set(s1)
print(s10)
```

### **1-3. Set 의 연산자**

```python
s1 = {10, 20, 30, 40, 50}
s2 = {30, 40, 50, 60, 70}

# 합집합(union) : union 또는 |
result = s1 | s2
print(result) # 더한 후, 중복값을 지움
result = s1.union(s2)
print(result)

# 차집합(difference) : -
result = s1 - s2
print(result)
result = s2.difference(s1)
print(result)

# 대칭 차집합(symmetric_difference(^)) : 합집합에서 교집합을 뺀 것 (A-B U B-A)
result = s1 ^ s2
print(result)
result = s1.symmetric_difference(s2)
print(result)
```
