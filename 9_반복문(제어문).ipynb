# **1. 대입 연산자**

* =, +=, -=, *=, /=, %=
* 왼쪽에 저장된 값과 오른쪽의 값을 연산한 뒤에 다시 왼쪽에 대입하는 연산자

````python
num = 10
num += 1
print(num) # 11
num -= 1
print(num) # 10
num *= 10
print(num) # 100
num /= 10
print(num) # 10
num %= 2 # 나머지 연산
print(num) # 0 
````

# **2. while 문**

```
while 조건식(boolean) :
  조건식이 참일 경우 문장을 실행함
  while 로 되돌아
```

````python
i = 1
while i < 5 :
  print(i, end='')
  i += 1
  print(' + 1 = {}'.format(i))
print('현재 i 의 값 : {}'.format(i))
while i > 1 :
  print(i, end='')
  i -= 1
  print(' - 1 = {}'.format(i))
print('현재 i 의 값 : {}'.format(i))
````

````python
# 1 ~ 10 까지의 합
i = 1
sum = 0

while i <= 10 :
  print("%d + %d = " % (i, sum), end='')
  sum += i
  i += 1
  print(sum)

print("1 ~ 10 까지의 합 : {}".format(sum))
````

# **3. for 문**

```
for 변수 in range(범위) :
  반복할 문장
  ...
```

````python
for i in range(0, 10, 1) : # i=0, i<=10, 1++
  print('{}'.format(i), end='')
````

````python
for i in range(10) : # default : 0부터 , x까지, 1씩증가
  print('{}'.format(i), end='')
````

````python
for i in range(0, 10) : # 생략해도 같은 결과가 나타남
  print('{}'.format(i), end='')
````

### 3-1. list 를 for 문과 함께 사용하기

````python
li = ['a', 'b', 'c', 'd']

for i in li : # i = a, i =  b, i =  c, i = d 가 각각 한번씩 실행
  print(i, end=' ')
````

# **4. 다중 반복문**

* for 또는 while 문이 2 개 이상 겹쳐있는 문

````python
for num1 in range(1, 4) :
  print('현재 num1 : {}'.format(num1))
  for num2 in range(1, 5) :
    print('현재 num2 : {}'.format(num2))
````
