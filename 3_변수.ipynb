# **1. 변수란?**

* 변수 : 처리할 데이터를 저장시키는 기억장소

```python
name = '김사과'
age = 20
print(name)
print(age,'\n')

name = '고세구'
print(name)
```

# **2. 변수의 이름을 작성하는 방법**

1. 영문자, 숫자, 특수문자(_) 로 사용하며 대소문자를 구분
2. 반드시 문자로 시작해야하며, *예약어는 사용할 수 없다 

  *예약어 : 문법에 사용하는 문자

```python
num1 = 10
num2 = 3
print('%d / %d = %.3f' % (num1, num2, num1 / num2))
```

# **3. 변수의 데이터 타입**

* 데이터 타입(자료형) : 자료가 저장되는 형태

```python
name = '고세구'
age = 20
weight = 50.5
isLover = False

#type() : 변수의 데이터타입을 반환해주는 함수

print(type(name))
print(type(age))
print(type(weight))
print(type(isLover)) # boolean 타입 (논리, true / false)

# C언어에서는 int a = 10 으로 타입을 지정, but python 에서는 a = 10 으로 지정하면 자동으로 변환하여 저장
```
