# **1. 튜플(Tuple)**

* 튜플은 리스트와 비슷하지만, 리스트는 [ ] 로 자료를 감싸고, 튜플은 ( ) 로 자료를 감쌈
* 튜플은 수정/삭제가 불가능함 # 상수
* 튜플은 기본적으로 수정/삭제 이외에는 리스트와 기능이 같다


```python
tu1 = () # 빈 튜플
print (tu1)

# tu1[0] = 10 # 'tuple' object does not support item assignment

tu2 = (1)
print(tu2)

tu3 = 1, 2, 3 # 소괄호() 를 생략해도 됨 (권장x)
print(tu3)

tu4 = ('apple', 'banana', ('strawberry', 'tomato')) # 이차원으로 생성 가능
print(tu4)
```

```python
print(type(tu4)) # <class 'tuple'>
```

# **2. 튜플의 수정과 삭제**

```python
tu5 = (1, 2, 'apple', 'banana')
print(tu5)

print(tu5[3])
# del tu5[3] # 'tuple' object doesn't support item deletion
del tu5 # 튜플 자체를 삭제하는 것은 가능, 내부의 멤버를 삭제하는 것은 불가능

tu6 = (1, 2, 'apple', 'banana')
print(tu6)

tu6 = 10 # 수정된 것이 아닌 가리키는 주소가 바뀐 것(포인터는 변경이 가능하다)
print(tu6)
```

튜플과 리스트는 기능이 같은데 그런데 튜플을 쓰는 이유 : 
* 변수와 상수의 개념, 변수를 생성하면 변수는 수정이 가능하고, 상수는 불가능
* 만약 값이 바뀌면 안되는 상수의 개념이 필요한 코드에서 임의로 값을 수정하지 못하도록 튜플을 이용하여 상수를 생

# **3. 튜플의 인덱싱, 슬라이싱, 연산**

```python
# 인덱싱
tu7 = (1, 2, 'apple', 'banana')
print(tu7[0])
print(tu7[-1]) # 리스트와 같다

# 슬라이싱
print(tu7[1:])
print(tu7[1:-1]) # -1은 포함하지 않음 = 리스트와 같다

# 덧셈(연결)
tu8 = (30, 40)
print(tu7 + tu8)

# 곱셈(반복)
print(tu8 * 3)

# 요소의 갯수
print(len(tu8)) # length
```
