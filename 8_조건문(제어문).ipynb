# **2. if 문**

```
if 조건식(결과가 True 또는 false) :

```

````python
age = int(input('나이를 입력하세요 : '))

if age >= 20 :
  print("성인")
  print("넷플릭스 성인 영화를 시청할 수 있습니다")
print("프로그램을 종료합니다")
````

# **3. if ~ else 문**

```
if 조건식(결과가 True 또는 false) :
  조건식이 True 일 경우 실행
else :
  조건식이 False 일 경우 실행
```

````python
age = int(input('나이를 입력하세요 : '))

if age >= 20 :
  print("성인")
  print("넷플릭스 성인 영화를 시청할 수 있습니다")
else :
  print("미성년자")
  print("디즈니를 보세요")
print("프로그램을 종료합니다")
````

# **4. if ~ elif ~ else 문**

```
if 조건식1(결과가 True 또는 false) :
  조건식1이 True 일 경우 실행
elif 조건식2 :
  조건식1이 False, 조건식2가 True 일 경우 실행
elif 조건식3 :
  조건식3이 True 일 경우 실행
...
else :
  모든 조건식이 False 일 경우 실행
```

````python
age = int(input('나이를 입력하세요 : '))

if age >= 20 :
  print("성인")
  print("넷플릭스 성인 영화를 시청할 수 있습니다")
elif age > 14 : # 위에서부터 내려오기 때문에 20보다 작다는 조건은 써줄 필요가 없다
  print("청소년")
elif age > 6 :
  print("어린이")
elif age > 0 :
  print("유아")
else :
  print("잘못 입력했습니다")
print("프로그램을 종료합니다")
````

### **문제**
국어점수, 수학점수, 영어점수를 한번에 입력받아 총점, 평균, 학점을 구하는 프로그램을 만든다.

(단, 학점은 평균 기준이며, 90점 이상은 A, 80점 이상은 B, 70점 이상은 C, 60점 이상은 D, 나머지는 F 학점이다.)

````python
kor, eng, math = map(int, input('시험점수를 입력하세요 : ').split(', '))

tot = kor + math + eng
avg = tot / 3
Hj = ""
if avg >= 90 :
  Hj = 'A'
elif avg >= 80 :
  Hj = 'B'
elif avg >= 70 :
  Hj = 'C'
elif avg >= 60 :
  Hj = 'D'
else :
  Hj = 'F'
print("총점 : ", tot)
print("평균 : ", avg)
print("학점 : ", Hj)
````

# **5. 논리 연산자**

### 5-1. AND 연산자
* 양쪽 두 식의 값이 모두 True 인 경우 결과가 True 가 되는 연산자

|  A  |  B  | 결과 |
|-----|-----|------|
|True |True |True  |
|True |False|False |
|False|True |False |
|False|False|False |

> 정보처리기사의 합격 기준은 모든 과목의 평균 점수가 60점 이상이고(and), 각 과목의 점수가 40점 이상이어야 함

````python
userid, userpw = map(str, input("아이디와 비밀번호를 입력하세요 : ").split(", "))
# map 함수에는 list 형밖에 사용하지 못함(split 사용하지 않으면 정상작동하지 않음)

if (userid == "admin") and (userpw == "1234") :
  print("로그인에 성공하였습니다")
elif userid == "admin" :
  print("비밀번호를 확인하세요")
elif userpw == 1234 :
  print("아이디를 확인하세요")
else :
  print("아이디와 비밀번호를 확인하세요")
````

### 5-2. OR 연산자
* 양쪽 두 식의 값중 하나라도 True 인 경우 결과가 True 가 되는 연산자

|  A  |  B  | 결과 |
|-----|-----|------|
|True |True |True  |
|True |False|True  |
|False|True |True  |
|False|False|False |

> 선생님이 엄마, 아빠 중 한분만 데려오라고 함

````python
# 주민등록번호에 의한 성별 판별
# 001011-3068518
jN = input("주민등록번호 13자리를 -없이 입력하세요 : ")
sung = ""

if jN[6] == "1" or jN[6] == "3" :
  sung = "남자"
elif jN[6] == "2" or jN[6] == "4" :
  sung = "여자"
else :
  sung = "알 수 없음"
print(sung)
````

### **문제**
주민등록번호를 입력받아 현재 나이를 구하는 프로그램을 작성한다

````python
jN = input("주민등록번호 13자리를 - 없이 입력하세요 : ") # jN = 주민번호를 입력받음
year = int(jN[:2])

if int(jN[6]) <= 2 : # 1900년대 사람이라면?
  year = year + 1900 # 93 + 1900 = 1993
else : 
  year = year + 2000 # 00 + 2000 = 2000

age = 2022 - year +1 # 한국식 나이 +1
print("현재나이는 {}살입니다".format(age))
````
