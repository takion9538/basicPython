# **1. 스페셜 메소드**

* __ 로 시작해서 __ 로 끝나는 특수함수
* 해당 메소드들을 구현하면, 커스텀 객체에서 여러가지 파이썬 내장 함수나 연산자를 적용 가능
* 오버라이딩 가능한 함수 목록은 아래 링크에서 참조
> https://docs.python.org/3/reference/datamodel.html


```python
class Point : 
  def __init__(self, x, y) :
    self.x = x
    self.y = y

  def print_pt(self) :
    print('({}, {})'.format(self.x, self.y))  

  def __str__(self) : # str() 를 재정의
    return ('({}, {})'.format(self.x, self.y))

  def __add__(self, other) :
    new_x = self.x + other.x
    new_y = self.y + other.y
    return Point(new_x, new_y)

  def __sub__(self, other) :
    new_x = self.x - other.x
    new_y = self.y - other.y
    return Point(new_x, new_y)

  def __mul__(self, other) :
    new_x = self.x * other
    new_y = self.y * other
    return Point(new_x, new_y)

  def length(self) :
    return self.x ** 2 + self.y ** 2

  def __len__(self) :
    return self.x ** 2 + self.y ** 2

  def __getitem__(self, index) :
#    selfarr = list(range(2))
#    selfarr[0] = self.x
#    selfarr[1] = self.y 
    if index == 0 :
      return self.x
    elif index == 1 :
      return self.y
    else :
      return self.y
```

```python
p1 = Point(3, 4)
print(p1)
p1.print_pt()
```

```python
# str() : 객체를 매개변수로 저장하면 객체의 주소 및 클래스를 확인
p2 = Point(5, 8)
print(p2) # print(str(p2)) 이 생략되어 있음 str 을 오버라이드했기 때문에 값이 달라짐
print(str(p2))
```

### **문제**

p1.x + p2.x = p3.x , p1.y + p2.y = p3.y
p1.x - p2.x = p4.x , p1.y - p2.y = p4.y

```python
# __add__ : + 를 override 하는 메소드

p3 = p1 + p2
print(p3)
```

```python
# __sub__ : - 를 override 하는 메소드

p4 = p1 - p2
print(p4)
```

```python
# __mul__ : * 를 override 하는 메소드

p5 = p1 *3
print(p5)
```

```python
# __len__

print(len(p1))
```
