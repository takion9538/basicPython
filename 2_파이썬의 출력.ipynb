매개변수 : 함수의 소괄호 안에 넣어 주는 값
따옴표를 써야하는 이유 : 컴퓨터는 문자를 이해하지 못하기 때문에 매개변수가 문자열이라는 것을 알려주기 위함
(따옴표 안에 들어간 값은 해석할 필요가 없다는 것을 알려줌)

```python
''' 
여기는 
여러줄 
주석문입니다
'''
"""
여기도
여러줄
주석문입니다
"""
#숫자 출력은 따옴표를 사용하지 않는다 한줄 주석문
print(100)
#문자 출력은 따옴표를 사용한다
print("python")
print('python')
```

```python
#띄어쓰기
print('Hello', 'Python')
print('Hello', 'Python', 'Hi', 'World')
print('Hello', end='') #end : 출력 후에 어떻게 처리할 지의 여부(옵션) -> hello 이후 다음 실행 내용을 뒤에 붙임
print('Python')
print('Hello', end=' ') #띄어쓰기가 들어갈 경우 띄어쓰기를 한 후 붙임
print('Python')
```

```python
print(10 + 5) # 덧셈
print(10 - 5) # 뺄셈
print('10' + '5') # 연결연산
# print('10' + 5) 에러 : Strint(10) + Int(5) 는 연산할 수 없음
print('10', 5)
print('10', end='')
```

# **2. 출력 서식 지정하기**

1. (%d : 정수, %f : 실수, %s : 문자) 를 사용하여 출력 서식을 지정
2. 서식을 제외한 문자는 입력한 그대로 출력
3. 서식을 지정해 출력할 데이터가 2개 이상일 경우 소괄호()로 데이터를 묶어야 함

```python
print('%d' % 100) # %뒤의 값을 정수형으로 출력 (decimal)
print('%d' % 100.1)
print('%f' % 168.7) # %뒤의 값을 실수형으로 출력 (float)
print('%.1f' % 168.7) # 소숫점 첫째자리까지 출력
print('%s' % '나는바보다') # %뒤의 값을 문자열로 출력 (String)
print('%s' % 150) # 숫자처럼 보여도 문자열임
```

```python
print('%5d' % 100) # decimal로 5자리를 확보한 후 100을 출력
#  100 이 출력된다
print('%5d' % 10000)
print('%05d' % 100)
```

```python
print('%6.2f' % 123.45) # %6(총 글자수).2(소숫점 둘째자리)
print('%6.2f' % 123.458) # 반올림
print('%6.2f' % 123.4) # 남은 자리는 0으로 채움
```

```python
print('%5s' % 'abc')
```

```python
# 5 + 3 = 8 을 출력
print('5 + 3 = 8')
print('%d + %d = %d' % (5, 3, 5 + 3))

# 5 * 3 = 15
print('%d * %d = %d' % (5, 3, 5 * 3))

# 5 / 3 = ?
print('%d * %d = %d' % (5, 3, 5 / 3)) # 결과를 정수로 출력(버림)
print('%d * %d = %.3f' % (5, 3, 5 / 3)) # 결과를 실수로 출력
```

# **3. format 함수와 출력 인덱스 사용**

index : 0부터 순서대로 진행되는 정수

```python
print('{0} {1}'.format('김사과', '학생')) # 문자열에 대한 format 을 지정
print('{1} {0}'.format('김사과', '학생')) # format(1,2,3)순서로 index가 생성되고, 앞에서 지정한 문자열에 대응된다
```

```python
print('{0:3s} 님은 {1:03d}살입니다'.format('김사과', 20))
print('{0} 님의 키는 {1:6.2f}cm 입니다'.format('아이네', 158.2))
```
