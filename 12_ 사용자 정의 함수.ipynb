````python
def funcq(*args) :
  max = args[0]
  min = args[0]
  for i in args :
    if max < i :
      max = i

    if min > i :
      min = i
  return max, min
````

````python
max, min = funcq(10, 90, 80, 50, 100, 80, 30)

print('ìµœëŒ“ê°’ : {}'.format(max))
print('ìµœì†Ÿê°’ : {}'.format(min))
````
