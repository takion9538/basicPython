### **1-1. 이름만 있는 함수(매개변수가 없는 함수)**
```
def 함수명() :
  함수가 호출되면 실행할 문장
```

````python
def func1() :
  print('Hello Python!')
````

````python
func1()
````

### **1-1. 매개변수가 있는 함수**
```
def 함수명(매개변수1, 매개변수2, ...) :
  함수가 호출되면 실행할 문장
```

````python
def func2(a, b) :
  for i in range(a) :
    print('입력받은 숫자 : %d' % b)

func2(2, 5)
````

````python
def func3(start, end) : 
  sum = 0
  for i in range(start, end + 1, 1) :
    sum += i
  print('{} 부터 {} 까지의 합은 {}'.format(start, end, sum))
````

````python
func3(5, 15)
````

### **1-3. 리턴값이 있는 함수(함수의 결과값을 다시 반환하는 함수)**
```
def 함수명(매개변수1, 매개변수2, ...) :
  함수가 호출되면 실행할 문장
  return 값
```

````python
def func4() : 
  return '👻'
````

````python
print(func4())
````

````python
a = func4()
print(a)
````

````python
def func5(num1, num2) :
  return num1 + num2
````

````python
result = func5(10, 5) # 값이 나오지 않고 함수에 반환되고, result 값에 저장됨
print(result) # print 를 통해 출력함
# return 을 사용하면 계산을 계속 하지 않고 한번 계산된 값을 변수에 저장하여, 연산이 빨라지는 장점이 있음
````

### **1-4. 디폴트 매개변수**
```
def 함수명(매개변수1=값1, 매개변수2=값2) :
  함수를 호출하면 실행할 문장
  return 
```

````python
def func6(num1=0, num2=0) : 
  return num1 + num2
````

````python
print(func6(10, 5))
print(func6()) # 0 + 0 = 0
print(func6(10)) # 10 + 0 = 0 : num1 = 10 으로 앞에서부터 순서대로 변수에 들어간다
````

### **1-5. 매개변수의 갯수가 가변적인 함수**
```
def 함수명(*매개변수) :
  함수가 호출되면 실행할 문장
  return 값
```
> 매개변수는 함수안에서 튜플형태로 사용하게 됨

````python
def func7(*args) : # 튜플의 형태로 저장
  result = 0
#  print(args) # 튜플의 형태를 확인할 수 있음 (10, 20, 30, 60, 40, 80)
  for i in args :
    result += i
  return result
````

````python
print(func7(10, 20, 30, 60, 40, 80))
print(func7(10, 20, 60, 40))
````

### **1-6. 매개변수를 딕셔너리 구조로 전달받는 함수**
```
def 함수명(**매개변수) :
  함수가 호출되면 실행될 문장
  return 값
```

````python
def func8(**args) : # 딕셔너리의 형태로 저장
  return args
````

````python
user = func8(userid='apple', name='아이네', age='27')
print(user)
````

### **1-7. 리턴값이 여러개인 함수**
```
def 함수명(매개변수1, 매개변수2, ...) :
  함수가 실행되었을 경우
  return 값, 값
```

````python
def func9(num1, num2) :
  return num1 + num2, num1 - num2, num1 * num2, num1 / num2
````

````python
result1, result2, result3, result4 = func9(10, 5)
print(result1, result2, result3, result4)
````

### **문제**

데이터 90, 80, 50, 100, 80 을 매개변수로 전달해서 최대값, 최소값을 출력하는 함수를 만들어 활용해보자   
(단, return 을 사용하고, 내장함수를 사용하지 않는다.)

````python
def funcq(*args) :
  max = args[0]
  min = args[0]
  for i in args :
    if max < i :
      max = i

    if min > i :
      min = i
  return max, min
````

````python
max, min = funcq(10, 90, 80, 50, 100, 80, 30)

print('최댓값 : {}'.format(max))
print('최솟값 : {}'.format(min))
````
