# **1. 문자열의 저장**

```python
str1 = '오늘은 3일차 파이썬 강의입니다'
str2 = "오늘은 3일차 파이썬 강의입니다"
str3 = '''오늘은 3일차 
파이썬 강의입니다''' # 줄바꿈이 포함된 문자열을 저장할 수 있다

print(str1)
print(str2)
print(str3)
```

# **2. 문자열의 인덱싱**

* 문자마다 번호를 매겨서 순서대로 사용하는 방법

```python
str = "Python can be easy to pick up whether you're a first time programmer or you're experienced with other languages."
# 문자열 안에 따옴표가 있을 경우에는 다른 따옴표로 묶어줄 수 있다
print(str)
print(type(str)) # 타입을 확인 <str>
```

```python
# 문자열의 인덱싱
print(str[0]) # P
print(str[1]) # y
print(str[20]) # o(easy t[o])
print(str[-1]) # , : 마지막 글자
print(str[-2]) # s : 뒤에서 두번째 글자
```

# **3. 문자열 슬라이싱**

* 001011 - 3068518 : 주민번호 예제

```python
str = "Python can be easy to pick up whether you're a first time programmer or you're experienced with other languages."
print(str[0:6]) # index {0} ~ {5} 까지의 문자열
print(str[14:22]) # {14} ~ {21} 까지의 문자열
print(str[22:]) # 인덱스 {22} ~ 끝까지의 문자열
print(str[:21]) # {0} ~ {20} 까지의 문자열
```

```python
str = "Python can be easy to pick up whether you're a first time programmer or you're experienced with other languages."
 
# 문자열 함수는 str. 후에 caplock 을 눌러서 나오는 함수
# len() : 인수로 입력된 문자열의 갯수를 반환 -> 파이썬의 내장 함수
print(len(str))

# count() : 문자열에서 인수로 입력된 문자열의 출현 횟수를 반환(내가 입력한 문자가(문자열이) 문자열에서 몇 개 들어있는지 센다)
print(str.count('s')) # 3
print(str.count('you')) # 2
print(str.count('YOU')) # 0 : 대소문자를 구분

# find() : 인수로 입력된 문자열이 처음 나오는 위치를 인덱스로 반환
print(str.find('you')) # 38
print(str.find('z')) # -1 : 찾는 문자열이 없을 경우 -1

# index() : 인수로 입력된 문자열이 처음 나오는 위치를 인덱스로 반환, 찾지 못했을 경우 error 발생
print(str.index('you')) # 38
# print(str.index('z')) # substring not found error

# replace() : 문자열을 치환
print(str.replace('Python', 'JavaScript'))
# print(str.replace('JavaScript', 'Python'))

# upper() : 대문자로 변환
print(str.upper())
print(str.lower())

str = '           Hello     Python          '
print(str)

# lstrip() : 문자열 왼쪽의 공백을 제거
print(str.lstrip())

# rstrip() : 문자열 오른쪽의 공백 제거
print(str.rstrip())

# strip() : 문자열 양쪽의 공백을 제거
print(str.strip())

str = "Python"

# join() : 문자열에 특정 문자열을 삽입
print('😘'.join(str))
```
