# **1. 파일 열기**

```
  파일 변수 = open('텍스트 파일 이름', '파일 열기 모드')
```

* 파일 열기 모드 : w(출력), a(추가), r(입력)
* w(write) 는 지정된 경로에 파일이 없을 경우 파일을 생성한 후 저장하고, 파일이 있으면 기존의 모든 데이터를 삭제하고 다시 저장함 (덮어쓰기)
* a(append) 는 지정된 경로에 파일이 없으면 파일을 생성한 후 저장하고, 파일이 있으면 기존의 데이터 뒤에 저장한다
* r(read) 은 지정된 경로에 파일이 없으면 에러가 발생함
* 출력시 (w, a) 작업이 완료되면 반드시 close() 메소드를 사용해서 파일을 닫아야 정상적으로 데이터가 저장되고, 입력시 파일을 닫지 않아도 정상적으로 처리됨

````python
file = open('data.txt', 'w')
for i in range(10) : 
  file.write('Hello, world' + str(i) + '\n')
file.close()
print('data.txt 파일에 쓰기 완료!!')
````

* 절대경로 : 디스크 드라이브의 최상위 (root) 디렉토리부터 파일이 저장된 디렉토리까지의 경로
  - D:\AI\Python\data.txt
* 상대경로 : 현재 화면에 표시되는 소스 파일이 위치한 폴더부터 테스트 파일이 저장될 폴더까지의 경로
  - data.txt (현재 폴더)  
  - ../data.txt (상위 폴더)  
  - ./data/data.txt (하위 폴더)

````python
file = open('./data/data.txt', 'a')
for i in range(10) : 
  file.write('Hello, world' + str(i) + '\n')
file.close()
print('data.txt 파일에 쓰기 완료!!')
````

### **문제**
키보드로 입력하는 문자열을 텍스트 파일에 저장 (단, 'quit' 라는 단어가 입력되면 저장을 끝내고, 프로그램을 종료함)

````python
file = open('./data/input.txt', 'w')

while True :
  text1 = input("문자열을 입력하세요 (quit = 종료) : ")
  if text1.lower() == 'quit' : # lower = 대소문자를 구분하지 않음(소문자로 변경)
    break
  file.write(text1 + '\n')

file.close()
print('input 파일 저장 완료!')
````

````python
with open('./data/input.txt', 'w') as file : 
  while True :
    text1 = input("문자열을 입력하세요 (quit = 종료) : ")
    if text1.lower() == 'quit' : # lower = 대소문자를 구분하지 않음(소문자로 변경)
      break
    file.write(text1 + '\n')
print('input 파일 저장 완료!')
````

# **2. 파일 읽기**

````python
file = open('./data/input.txt', 'r')
lines = []

while True : 
  line = file.readline()
  if not line : 
    break
  if len(line.strip()) != 0 : # 공백을 지움
    print(line, end = '')
    lines.append(line.strip())
file.close()
print('data.txt 파일 읽기 완료')
print(lines)
````

````python
file = open('./data/input.txt', 'r')

lines = file.readlines()
print(type(lines)) # 리스트 타입으로 데이터를 받아온다
print(lines)

file.close()
print('input.txt 파일 읽기 완료')
print(lines)
````
