# **1. 딕셔너리(Dictionary)**

* 대응관계를 나타내는 자료형으로, key - value 라는 것을 한 쌍으로 갖는다
* 하나의 딕셔너리의 key 값은 중복될 수 없다
* 하나의 딕셔너리의 값은 중복될 수 있다.

### **1-1. 딕셔너리 생성**

```python
dic1 = {}
print(dic1)

dic2 = {'no':1, 'userid':'apple', 'name':'김사과', 'hp':'010-1111-1111'} # {key:value, key:value}
print(dic2)
```

### **1-2. 키 를 통해 값 찾기**

```python
print(dic2['no'])
print(dic2['userid'])
print(dic2['name'])
print(dic2['hp'])
```

### **1-3. 데이터 추가 및 삭제**

```python
dic3 = {1:'apple'}
print(dic3)

dic3[100] = 'orange'
print(dic3)

dic3[50] = 'banana'
print(dic3)

dic3[1] = 'melon' # 키 값은 중복될 수 없기 때문에 변경됨
print(dic3)

del dic3[50]
print(dic3)
```

### **1-4. 딕셔너리 함수**

```python
# keys() : key 리스트를 반환 *
print(dic2.keys())

# values() : value 리스트를 반환
print(dic2.values())

# items() : key 와 value 를 한 쌍으로 묶는 튜플을 리턴
print(dic2.items()) # ('no', 1)

# get() : key 를 이용해서 value 를 반환 *
print(dic2.get('name'))
print(dic2.get('hp'))
print(dic2.get('age')) # none
print(dic2['no']) # 함수를 쓰지 않고 인덱싱을 통해 가져올 수 있지만, 함수를 이용하면 파라미터를 통해 가져오기 때문에 매개변수를 이용한 기능을 사용할 수 있는 이점이 있다
print(dic2.get('age', '모름')) # 값이 None 일 때 '모름' 을 출력함

# in : key 가 딕셔너리 안에 있는지 확인하고 boolean 으로 반환
print('name' in dic2) # True
print('age' in dic2) # False
```

### **1-5. 딕셔너리와 for**

```python
for i in dic2 : # i 값에 key 가 저장된다 = dic2.keys()
  print(i, end=' ')
print("\n")

for i in dic2.keys() : 
  print(i, end=' ')
print('\n')

for i in dic2.values() :
  print(i, end=' ')
print('\n')

for i in dic2 : # i 값을 dic 에 인덱싱 하는 방법으로 사용할 수 있다
  print(dic2[i], end=' ')
print('\n')
```

**과제**

아래와 같은 딕셔너리 자료에서 주민등록번호를 추출하여 성별을 알아내는 프로그램을 작성
(단, ssn의 7번째 자리의 성별은 1, 3 은 남자, 2, 4 는 여자 라고 함)

> user = {'name':'김사과', 'hp':'010-1111-1111', 'ssn':'0010114068518'}

```python
user = {'name':'김사과', 'hp':'010-1111-1111', 'ssn':'0010114068518'}

star = user['ssn']
print(star)
a = star[6]
print(a)
if a == '1' or a == '3' :
  print('남자')
elif a == '2' or a == '4' :
  print('여자')
```

```python
user = {'name':'김사과', 'hp':'010-1111-1111', 'ssn':'0010114068518'}

for i in user.keys() :
  if i == 'ssn' :
    ssn = user.get(i) # 0010114068518
    if ssn[6] == '1' or ssn[6] == '3':
      print('남자')
    elif ssn[6] == '2' or ssn[6] == '4' :
      print('여자')
```
