# **1. 리스트(List)**

* 여러가지 데이터를 콤마(,) 로 구분하여 대괄호[ ]로 감싼 자료구조 = 배열

```python
li1 = [1, 3, 5, 7]
print(li1)

li2 = list([1, 3, 5])
print(li2)

li3 = ['Python', 'C++', 'JavaScript', 'Java']
print(li3)

# *여러개의 데이터타입을 한꺼번에 저장할 수 있다
li4 = [1, 1.5, 'Python', True]
print(li4)
print(type(li4),type(li4[0]), type(li4[1]), type(li4[2]), type(li4[3]))

li5 = [1, 2, 'Python', ['apple', 'banana']] # 리스트 안에 리스트를 포함할 수 있다
print(li5[3])
print(li5[3][1])
```

# **2. 리스트 인덱싱**

```python
li1 = [1, 3, 5, 7]
print(li1[0])
print(li1[-1])
print(li1[0] + li1[2])
```

```python
# 다중 리스트
li2 = [1, 3, 5, ['😘', '😢', '💕']]
print(li2[0])
print(li2[-1])
print(li2[-1][2])
```

```python
li3 = [1, 3, 5, ['apple', 'banana', ['🥲', '😋']]]
print(li3[2])
print(li3[-2])
print(li3[-1])
print(li3[-1][-2])
print(li3[-1][-1][-1])
```

# **3. 리스트 슬라이싱**

```python
li4 = [10, 20, 30, 40, 50]
print(li4)
print(li4[0:3])
print(type(li4)) # 자료구조는 list

# li5 = li4 이면 li4 가 가리키고 있는 [10, 20, 30, 40, 50] 의 주소를 가져온다
li5 = li4 # 둘이 가리키는 배열이 같다
print(li5)
li5[0] = 100 # li5 가 가리키는 배열이 li4 가 가리키는 배열과 같기 때문에 둘의 내용 모두가 수정된다
print(li5) # li5[0] = 100
print(li4) # li4[0] = 100

li6 = [10, 20, 30, ['apple', 'banana', 'orange'], 40, ['😋', '💕']]
print(li6[2:6]) # == li6[2:]
print(li6[5][:1])
```

# **4. 리스트 연산자**

```python
li7 = [10, 20, 30]
li8 = [40, 50, 60]

print(li7 + li8) # 두 개의 리스트가 합쳐진다 [10, 20, 30, 40, 50, 60]
print(li8 + li7) # 바뀐 상태로 합쳐진다(문자열의 덧셈) [40, 50, 60, 10, 20, 30]
print(li7 * 3) # 문자열과 같다 [10, 20, 30, 10, 20, 30, 10, 20, 30]
print('😋' * 10) # 10 번 반복된다

li9 = [10, 20, 30]
# print(li9[0] + '😋') # li9[0] 은 숫자이기 때문에 숫자와 문자열의 덧셈은 불가능 unsupported operand type(s) for +: 'int' and 'str'
li9[1] = '😋'
print(li9)
print('10' + '😋')
```

# **5. 리스트의 수정 및 삭제**

```python
# 리스트의 수정
li10 = [10, 20, 30]
li10[1] = 100
print(li10)

li10 = [10, 20, 30]
li10[1:2] = ['😋', '💕', '😍'] # 리스트를 슬라이싱으로 추가
print(li10) # 리스트 사이에 멤버(데이터)가 추가됨

li10 = [10, 20, 30]
li10[1] = ['😋', '💕', '😍'] # 리스트를 인덱싱으로 추가
print(li10) # 리스트 속에 리스트가 추가됨

print(li10[0:3])
li10[1:3] = [] # 데이터를 삭제
print(li10)

del li10[0] # del 명령을 이용하여 데이터삭제
print(li10)
```

# **6. 리스트 함수**

```python
# 리스트 객체를 생성하면서 사용할 수 있게 되는 함수
li11 = [10, 20, 30]

# append() : 리스트(뒤)에 데이터를 추가
li11.append(100)
print(li11)

li11.append([40, 50])
print(li11)
```

```python
li12 = [10, 20, 30, 40, 50]

# pop() : 리스트의 마지막에 저장된 데이터를 반환하고, 해당 데이터는 삭제한다
print(li12)
# print(li12.pop()) # 반환할 때 데이터를 찍어주고 사라짐
temp = li12.pop() # 반환한 데이터를 temp 에 저장
print(temp)
print(li12)
```

```python
li13 = [10, 20, 30, 10, 20, 80]

# remove() : 매개변수로 전달된 데이터를 제거, 여러개가 있다면 처음 출현하는 데이터를 제거
li13.remove(20)
print(li13)

li13.remove(20)
print(li13)
```

```python
li14 = [10, 20, 30]

# insert() : 리스트의 특정 위치에 데이터를 삽입
li14.insert(1, 100) # 1번 자리에 100이라는 데이터를 삽입
print(li14)
```

```python
li15 = [10, 20, 30, 40, 50]

# index() : 리스트에 저장된 데이터가 있을 경우 위치를 반환(없다면 error)
print(li15.index(30)) # 2
# print(li15.index(60)) # is not in list error 발생
```

```python
li16 = ['😍', '😋', '😘', '❤️', '🤣']

# reverse() : 리스트에 저장된 데이터의 순서를 반대로 재조합
li16.reverse()
print(li16)
```

```python
li17 = [10, 40, 30, 100, 90, 50]

# sort() : 리스트에 저장된 데이터를 정렬
li17.sort() # 오름차순
print(li17)
li17.sort(reverse=True) # 내림차순
print(li17)

li18 = ['Apple', 'apple', 'orange', 'banana', 'melon']
li18.sort()
print(li18) # 대문자가 먼저 정렬된다 -> 아스키 코드표 순서
```

```python
li19 = [10, 20, 30, 40, 50, 20, 40, 30, 20]

# count() : 리스트에 포함된 특정 데이터의 갯수를 반환
print(li19.count(20)) # 3
print(li19.count(15)) # 0
```

```python
li20 = [1, 2, 3]

# extend() : 리스트를 추가(리스트에 리스트를 더함)
li20.extend([40, 50])
print(li20)

# list 의 덧셈과 같다
li21 = [1, 2, 3]
li21 = li21 + [40, 50] # li21 += [40, 50] 과 같다
print(li21)

# append 와 같은 기능이지만, append 는 하나의 객체만 추가가 가능
```
