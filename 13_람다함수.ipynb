# **1. 람다함수**

* 단일문으로 표현되는 익명함수
* 익명함수란 이름이 없는 구현체만 존재하는 간단한 함수를 의미한다
* 코드상에서 한 번만 사용되는 기능이 있을 때, 굳이 함수로 만들지 않고 일회성으로 만들어 사용할 때 사용하는 것
* 굳이 사용하지 않아도 되지만 코드가 가독성있고, 짧게 요약하기 위해 사용

```python
def square1(x) :
  return x ** 2 # ** = 제곱

print(square1(10))
```

```python
square2 = lambda x : x ** 2 # square2 변수를 람다함수로 이용

print(square2(5))
```

```python
def add1(x, y) :
  return x + y

print(add1(10, 5))
```

```python
add2 = lambda x, y : x + y

print(int(add2(10, 5)))
```

```python
strings = ['aaaa', 'bbb', 'cc', 'ddddd']
strings.sort(key = lambda s : len(s)) # 람다함수가 여러번 for 과 같이 실행됨
print(strings)
```

# **2. filter**

* 람다가 유용하게 사용되는 대표적인 함수
* 특정 조건을 만족하는 요소만 남기고 필터링

```python
def even(n) : 
  return n % 2 == 0 # 짝수라면 true 를, 홀수라면 false 를 return

print(even(3))
print(even(100))
```

```python
nums = [1, 2, 3, 4, 6, 9, 10, 14, 15, 18, 20]

print(list(filter(lambda n : n % 2 == 0, nums))) # false 를 필터링하고 출력
```

# **3. map**

* 함수형 프로그래밍의 기본 요소
* 각 원소를 주어진 수식에 따라 변형하여 새로운 리스트를 반환

```python
nums = [1, 2, 3, 4, 6, 9, 10, 14, 15, 18, 20]

print(list(map(lambda n : n ** 2, nums))) # 각 요소에 람다식을 적용하여 출력
```

```python
nums = [1, 2, 3, 4, 6, 9, 10, 14, 15, 18, 20] # 리스트 함수는 리스트 외에도 튜플에도 사용할 수 있다

print(list(filter(lambda n : n % 2 == 0, map(lambda n : n ** 2, nums))))
```
